## Evaluate atmospheric variables

```python
def compute_accuracy_and_range(results_temp):
    accuracy_iterations = []
    for _ in range(30):  # Adjust the number of iterations as needed
        sample_indices = np.random.choice(results_temp.index, size=int(0.33*len(results_temp)), replace=False)
        sampled_data = results_temp.loc[sample_indices]
        accuracy_iterations.append(accuracy_score(sampled_data['y_true'], sampled_data['y_predicted']))
    accuracy_mean = np.mean(accuracy_iterations)
    accuracy_std = np.std(accuracy_iterations)
    error_range = 1. * accuracy_std  # Assuming a 95% confidence interval
    return accuracy_mean, error_range

def process_results(indices, list_vars, path_results):
    all_accs = []
    all_ranges = []
    weeks = np.arange(1, 9)
    
    for ivar in range(len(indices)):
        name_temp = list_vars[indices[ivar]]
        acc_temp = []
        acc_range_temp = []
        
        for week in weeks:
            results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv', index_col=0, parse_dates=True)
            acc_mean, acc_range = compute_accuracy_and_range(results_temp)
            acc_temp.append(acc_mean)
            acc_range_temp.append(acc_range)
        
        all_accs.append(acc_temp)
        all_ranges.append(acc_range_temp)
    
    ind_max_acc = np.argmax(np.array(all_accs), axis=0)
    best_model = ['\n'.join(list_vars[indices[ind]].split('_')[:-1]) for ind in ind_max_acc]
    
    max_acc_range = [np.array(all_ranges)[ind_max_acc[i], i] for i in range(len(ind_max_acc))]
    
    best_model_all = best_model
    best_acc_all = np.max(np.array(all_accs), axis=0) * 100
    best_acc_range_all = np.array(max_acc_range) * 100
    mean_acc_all = np.mean(np.array(all_accs), axis=0) * 100
    diff_acc_all = best_acc_all - mean_acc_all
    
    return all_accs, all_ranges, best_model_all, best_acc_all, best_acc_range_all, mean_acc_all, diff_acc_all
    
def plot_accuracy(weeks, all_accs, best_model, acc_persistence, list_vars, component, indices_component):
    if component == 'lnd':
        component_long = 'land'
    if component == 'atm':
        component_long = 'atmosphere'
    if component == 'ocn':
        component_long = 'ocean'

    list_vars = np.array(list_vars)
    plt.figure(figsize=(8, 6))

    # Plot individual models
    for acc_temp, name_temp in zip(all_accs, [list_vars[indices_component][i].split('_')[:-1] for i in range(len(indices_component))]):
        plt.plot(weeks, np.array(acc_temp) * 100, label='_'.join(name_temp), alpha=0.4)

    # # Plot all variables for the given component
    # acc_temp = [accuracy_score(pd.read_csv(f'{path_results}results_{component}_week{week}.csv', index_col=0, parse_dates=True)['y_true'],
    #                            pd.read_csv(f'{path_results}results_{component}_week{week}.csv', index_col=0, parse_dates=True)['y_predicted'])
    #             for week in weeks]
    # plt.plot(weeks, np.array(acc_temp) * 100, label=f'All {component_long} variables', color='k', ls='--')

    # Plot best model
    for i, j, d in zip(weeks, np.max(np.array(all_accs), axis=0) * 100, best_model):
        plt.scatter(i, j, color='k', marker='*', s=100)
        plt.text(i - 0.25, j + 0.45, s=d)

    # Plot persistence
    plt.plot(weeks, acc_persistence * 100, color='red', label='Persistence', ls='dashdot', lw=2)

    # Finalizing plot
    plt.ylim(18, 37)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(alpha=0.2)
    plt.ylabel('Accuracy [%]',fontsize=14)
    plt.xlabel('Lead time [weeks]',fontsize=14)
    plt.title(f'Models based on {component_long} variables', loc='left')
    plt.legend(bbox_to_anchor=(1, 1.0), fontsize=14, ncol=2)
    plt.savefig(f'figures/acc_{component}.png', bbox_inches='tight')
    plt.show()
```

```python
# Call the function with necessary arguments for ocean
all_accs_ocean, all_ranges_ocean, best_model_ocean, best_acc_ocean, best_acc_range_ocean, mean_acc_ocean, diff_acc_ocean = process_results(indices_ocean, list_vars, path_results)

# Call the function with necessary arguments for land
all_accs_land, all_ranges_land, best_model_land, best_acc_land, best_acc_range_land, mean_acc_land, diff_acc_land = process_results(indices_land, list_vars, path_results)

# Call the function with necessary arguments for atmosphere
all_accs_atmosphere, all_ranges_atmosphere, best_model_atmosphere, best_acc_atmosphere, best_acc_range_atmosphere, mean_acc_atmosphere, diff_acc_atmosphere = process_results(indices_atmosphere, list_vars, path_results)
```

```python
indices_seasons = {'Winter':[12,1,2],
                   'Spring':[3,4,5],
                   'Summer':[6,7,8],
                   'Fall':[9,10,11]}

acc_persistence_seasons = {}
for season in indices_seasons.keys():
    acc_persistence_temp = []
    for wk in range(1,9):
        df_temp = df_shifts[[f'week{0}',f'week{wk}']].dropna()
        df_temp = df_temp[(df_temp.index.month==indices_seasons[season][0])|\
                                    (df_temp.index.month==indices_seasons[season][1])|\
                                    (df_temp.index.month==indices_seasons[season][2])]
        acc_persistence_temp.append(accuracy_score(df_temp[f'week{wk}'],df_temp[f'week{0}']))
    acc_persistence_seasons[season] = acc_persistence_temp
```

```python
def process_results_with_seasons(indices, list_vars, path_results, indices_seasons, season):
    all_accs = []
    all_ranges = []
    weeks = np.arange(1, 9)
    
    for ivar in range(len(indices)):
        name_temp = list_vars[indices[ivar]]
        acc_temp = []
        acc_range_temp = []
        
        for week in weeks:
            results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv', index_col=0, parse_dates=True)
            results_temp = results_temp[(results_temp.index.month==indices_seasons[season][0])|\
                            (results_temp.index.month==indices_seasons[season][1])|\
                            (results_temp.index.month==indices_seasons[season][2])]
            
            acc_mean, acc_range = compute_accuracy_and_range(results_temp)
            acc_temp.append(acc_mean)
            acc_range_temp.append(acc_range)
        
        all_accs.append(acc_temp)
        all_ranges.append(acc_range_temp)
    
    ind_max_acc = np.argmax(np.array(all_accs), axis=0)
    best_model = ['\n'.join(list_vars[indices[ind]].split('_')[:-1]) for ind in ind_max_acc]
    
    max_acc_range = [np.array(all_ranges)[ind_max_acc[i], i] for i in range(len(ind_max_acc))]
    
    best_model_all = best_model
    best_acc_all = np.max(np.array(all_accs), axis=0) * 100
    best_acc_range_all = np.array(max_acc_range) * 100
    mean_acc_all = np.mean(np.array(all_accs), axis=0) * 100
    diff_acc_all = best_acc_all - mean_acc_all
    
    return all_accs, all_ranges, best_model_all, best_acc_all, best_acc_range_all, mean_acc_all, diff_acc_all
    
```

```python
# Call the function with necessary arguments for each season
for season in indices_seasons.keys():
    best_accs = {}
    best_acc_ranges = {}
    best_names = {}
    component_names = ['atmosphere','ocean','land']
    for indices_component, component in zip([indices_atmosphere, indices_ocean, indices_land], component_names):
        all_accs_temp, all_ranges_temp, best_model_all, best_acc_all, \
            best_acc_range_all, mean_acc_all, diff_acc_all = \
                process_results_with_seasons(indices_component, list_vars, path_results, indices_seasons, season)
        best_accs[component] = best_acc_all
        best_acc_ranges[component] = best_acc_range_all
        best_names[component] = best_model_all
        print(best_names[component])

    plt.figure(figsize=(8, 7))
    
    plt.plot(weeks,best_accs['atmosphere'],label=f'Best atmosphere model',color='mediumorchid',lw=2)
    plt.scatter(weeks,best_accs['atmosphere'],color='mediumorchid')#, s=diff*100)
    plt.plot(weeks,best_accs['ocean'],label=f'Best ocean model',color='darkblue',lw=2)
    plt.scatter(weeks,best_accs['ocean'],color='darkblue')
    plt.plot(weeks,best_accs['land'],label=f'Best land model',color='darkorange',lw=2)
    plt.scatter(weeks,best_accs['land'],color='darkorange')
    
    plt.fill_between(weeks, best_accs['atmosphere'] - best_acc_ranges['atmosphere'], 
                     best_accs['atmosphere'] + best_acc_ranges['atmosphere'], color='mediumorchid', alpha=0.05)
    plt.fill_between(weeks, best_accs['ocean'] - best_acc_ranges['ocean'], 
                     best_accs['ocean'] + best_acc_ranges['ocean'], color='darkblue', alpha=0.05)
    plt.fill_between(weeks, best_accs['land'] - best_acc_ranges['land'], 
                     best_accs['land'] + best_acc_ranges['land'], color='darkorange', alpha=0.05)

    # Plot best model
    for i, j, d in zip(weeks, best_accs['atmosphere'], best_names['atmosphere']):
        plt.text(i - 0.25, 18, s=d, color = 'mediumorchid',fontsize=11)
    for i, j, d in zip(weeks, best_accs['ocean'], best_names['ocean']):
        plt.text(i - 0.25, 17, s=d, color = 'darkblue',fontsize=11)
    for i, j, d in zip(weeks, best_accs['land'], best_names['land']):
        plt.text(i - 0.25, 15.2, s=d, color = 'darkorange',fontsize=11)

    
    plt.plot(weeks, np.array(acc_persistence_seasons[season])*100, color='red', label='Persistence', ls='dashdot', lw=2)
    plt.xlim(0.4, 8.6)
    plt.ylim(15, 40)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(alpha=0.2)
    plt.ylabel('Accuracy [%]',fontsize=14)
    plt.xlabel('Lead time [weeks]',fontsize=14)
    plt.legend(bbox_to_anchor=(1, 1.0), fontsize=14)
    
    plt.title(f'{season}',loc='left',fontsize=14)
    # plt.legend(bbox_to_anchor=(1, 0.25))
    plt.legend(fontsize=14)
    
    plt.savefig(f'figures/acc_3comps_{season}.png',bbox_inches='tight')
    plt.show()
    plt.close('all')
    
        
```

# Changes in predictability depending on WR

```python
names_wr = ['WR1: AR','WR2: PT', 'WR3: GH', 'WR4: CUSH', 'WR5: Clima']
```

```python
acc_persistence_wrs = {}
for iwr,wr in enumerate(names_wr):
    acc_persistence_temp = []
    for wk in range(1,9):
        df_temp = df_shifts[[f'week{0}',f'week{wk}']].dropna()
        df_temp = df_temp[(df_temp[f'week{wk}'] == iwr)|(df_temp[f'week0'] == iwr)]
        acc_persistence_temp.append(accuracy_score(df_temp[f'week{0}'],df_temp[f'week{wk}']))
    acc_persistence_wrs[wr] = acc_persistence_temp
```

```python
def process_results_with_wr(indices, list_vars, path_results, wr_temp):
    all_accs = []
    all_ranges = []
    weeks = np.arange(1, 9)
    
    for ivar in range(len(indices)):
        name_temp = list_vars[indices[ivar]]
        acc_temp = []
        acc_range_temp = []
        
        for week in weeks:
            results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv', index_col=0, parse_dates=True)
            results_temp = results_temp[(results_temp['y_true']==wr_temp)|(results_temp['y_predicted']==wr_temp)]
            
            acc_mean, acc_range = compute_accuracy_and_range(results_temp)
            acc_temp.append(acc_mean)
            acc_range_temp.append(acc_range)
        
        all_accs.append(acc_temp)
        all_ranges.append(acc_range_temp)
    
    ind_max_acc = np.argmax(np.array(all_accs), axis=0)
    best_model = ['\n'.join(list_vars[indices[ind]].split('_')[:-1]) for ind in ind_max_acc]
    
    max_acc_range = [np.array(all_ranges)[ind_max_acc[i], i] for i in range(len(ind_max_acc))]
    
    best_model_all = best_model
    best_acc_all = np.max(np.array(all_accs), axis=0) * 100
    best_acc_range_all = np.array(max_acc_range) * 100
    mean_acc_all = np.mean(np.array(all_accs), axis=0) * 100
    diff_acc_all = best_acc_all - mean_acc_all
    
    return all_accs, all_ranges, best_model_all, best_acc_all, \
        best_acc_range_all, mean_acc_all, diff_acc_all
```

```python
col_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_oni = pd.read_csv('https://psl.noaa.gov/data/correlation/oni.data', sep='\s+', skiprows=1, names=col_names) 
df_oni = df_oni.apply(pd.to_numeric, errors='coerce')
df_oni = df_oni.iloc[:-9]
# df_oni=df_oni.dropna(inplace=True)
# df_oni=df_oni.replace({-99.90 : np.NaN}, inplace=True)
stacked_series = df_oni.stack()
df_oni_stacked = stacked_series.to_frame()
df_oni_stacked.reset_index(inplace=True)
df_oni_stacked.columns = ['Year', 'Month', 'ONI']
df_oni_stacked['Date'] = pd.to_datetime(df_oni_stacked['Year'] + '-' + df_oni_stacked['Month'])
df_oni_stacked['Date'] = df_oni_stacked['Date'].dt.strftime('%Y-%m')
df_oni_stacked.set_index('Date', inplace=True)
df_oni_stacked.drop(['Year', 'Month'], axis=1, inplace=True)
df_oni_stacked.index = pd.to_datetime(df_oni_stacked.index)

serie_oni = []
for i in range(len(df_shifts.index)):
    date_temp = df_shifts.index[i]
    serie_oni.append(df_oni_stacked.loc[f'{date_temp.year}-{str(date_temp.month).zfill(2)}']\
            .values[0][0])
    
df_shifts['ONI'] = serie_oni
df_oni = df_shifts[['ONI']]
```

```python
names_states = ['Niño','Neutral','Niña']

acc_persistence_enso = {}
for state in names_states:
    acc_persistence_temp = []

    for wk in range(1,9):
        df_temp = df_shifts[[f'week{0}',f'week{wk}']].dropna()
        
        df_temp = pd.concat([df_temp,df_oni],axis=1).dropna()
        if state == 'Niño':
            df_temp = df_temp[df_temp['ONI']>=0.5]
        elif state == 'Neutral':
            df_temp = df_temp[(df_temp['ONI']>-0.5)&(df_temp['ONI']<0.5)]
        elif state == 'Niña':
            df_temp = df_temp[df_temp['ONI']<=-0.5]
        else:
            break
            
        acc_persistence_temp.append(accuracy_score(df_temp[f'week{0}'],df_temp[f'week{wk}']))
        
    acc_persistence_enso[state] = acc_persistence_temp
```

```python
weeks = np.arange(1,9)
for state in names_states:
    for indices_component, component, label, color in zip([indices_atmosphere, indices_ocean, indices_land],
                                ['atm','ocn','lnd'],
                                ['atmosphere','ocean','land'],
                                ['mediumorchid', 'darkblue', 'orange']):
        
        all_accs = []
        for ivar in range(len(indices_component)):
            name_temp = list_vars[indices_component[ivar]]
            acc_temp = []
            for week in weeks:
                results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                          index_col=0,parse_dates=True)
                
                results_temp = pd.concat([results_temp,df_oni],axis=1).dropna()
                if state == 'Niño':
                    results_temp = results_temp[results_temp['ONI']>=0.5]
                elif state == 'Neutral':
                    results_temp = results_temp[(results_temp['ONI']>-0.5)&(results_temp['ONI']<0.5)]
                elif state == 'Niña':
                    results_temp = results_temp[results_temp['ONI']<=-0.5]
                else:
                    break
                acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
            all_accs.append(acc_temp)
        
        
        acc_temp = []
        for week in weeks:
            results_temp = pd.read_csv(f'{path_results}results_{component}_week{week}.csv',
                                              index_col=0,parse_dates=True)
            results_temp = pd.concat([results_temp,df_oni],axis=1).dropna()
            if state == 'Niño':
                results_temp = results_temp[results_temp['ONI']>=0.5]
            elif state == 'Neutral':
                results_temp = results_temp[(results_temp['ONI']>-0.5)&(results_temp['ONI']<0.5)]
            elif state == 'Niña':
                results_temp = results_temp[results_temp['ONI']<=-0.5]
            else:
                break
            acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
        # plt.plot(weeks,np.array(acc_temp)*100,label=f'All {label} variables',color=color,ls='--',alpha=0.4)
        all_accs.append(acc_temp)
        plt.plot(weeks,np.max(np.array(all_accs),axis=0)*100,label = f'Best {label} model',color=color)
        
        ind_max_acc = np.argmax(np.array(all_accs),axis=0)
        best_model = []
        for ind in ind_max_acc:
            if ind==len(all_accs)-1:
                best_model.append('ALL')
            else:
                best_model.append(' '.join(list_vars[indices_component[ind]].split('_')[:-1]))
        for i,j,d in zip(weeks,np.max(np.array(all_accs),axis=0)*100,best_model):
            plt.scatter(i,j,color=color)
            plt.text(i-0.25,j+0.27,s=d,color=color)
        
        plt.plot(weeks, np.array(acc_persistence_enso[state][:])*100,
             color='red',label='Persistence',ls='dashdot',lw=2)
    # plt.xlim(1,8)
    plt.ylim(19,40)
    plt.grid(alpha=0.3)
    plt.ylabel('Accuracy [%]')
    # plt.xlabel('Duration (weeks)')
    plt.title(f'Accuracy from each component - {state}: ',loc='left')
    plt.savefig(f'figures/acc_3comps_{state}.png'.replace('ñ','n'),bbox_inches='tight')
    # plt.legend(bbox_to_anchor=(1, 0.25))
    plt.show()
    plt.close('all')
```
