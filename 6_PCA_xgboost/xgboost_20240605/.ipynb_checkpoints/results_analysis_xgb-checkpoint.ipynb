```python
wr_series = pd.read_csv('/glade/work/jhayron/Data4Predictability/WR_Series_20231120.csv',\
                index_col=0,names=['week0'],skiprows=1,parse_dates=True)
for wk in range(2,10):
    series_temp = copy.deepcopy(wr_series["week0"])
    series_temp.index = series_temp.index - timedelta(weeks = wk-1)
    series_temp.name = f'week{wk-1}'
    if wk==2:
        df_shifts = pd.concat([pd.DataFrame(wr_series["week0"]),pd.DataFrame(series_temp)],axis=1)  
    else:
        df_shifts = pd.concat([df_shifts,pd.DataFrame(series_temp)],axis=1)
```

```python
all_accs = []
weeks = np.arange(1,9)
for ivar in range(len(indices_atmosphere)):
    name_temp = list_vars[indices_atmosphere[ivar]]
    acc_temp = []
    for week in weeks:
        results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                  index_col=0,parse_dates=True)
        acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
    plt.plot(weeks,np.array(acc_temp)*100,label=name_temp,alpha=0.4)
    all_accs.append(acc_temp)
component = 'atm'
acc_temp = []
for week in weeks:
    results_temp = pd.read_csv(f'{path_results}results_{component}_week{week}.csv',
                                      index_col=0,parse_dates=True)
    acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
plt.plot(weeks,np.array(acc_temp)*100,label='All atmosphere variables',color='k',ls='--')
all_accs.append(acc_temp)

ind_max_acc = np.argmax(np.array(all_accs),axis=0)
best_model = []
for ind in ind_max_acc:
    if ind==len(all_accs)-1:
        best_model.append('ALL')
    else:
        best_model.append('\n'.join(list_vars[indices_atmosphere[ind]].split('_')[:-1]))
# plt.plot(weeks,np.max(np.array(all_accs),axis=0)*100,label = 'Best atmosphere model',color='k')
for i,j,d in zip(weeks,np.max(np.array(all_accs),axis=0)*100,best_model):
    if i==1:
        plt.scatter(i,j,color='k',label = 'Best atmosphere model',
                    marker='*',s=100)
    else:
        plt.scatter(i,j,color='k',
                    marker='*',s=100)
    plt.text(i-0.25,j+0.45,s=d)
    
plt.plot(weeks, acc_persistence*100,color='red',label='Persistence',ls='dashdot',lw=2)

plt.ylim(18,37)
plt.grid(alpha=0.3)
plt.ylabel('Accuracy [%]')
plt.xlabel('Lead time [weeks]')
plt.title('a) Models based on atmospheric variables: ',loc='left')
plt.legend(bbox_to_anchor=(1, 1.0),fontsize=9)
plt.savefig('figures/acc_atm.png',bbox_inches='tight')

best_model_atmosphere = best_model
best_acc_atmosphere = np.max(np.array(all_accs),axis=0)*100
mean_acc_atmosphere = np.mean(np.array(all_accs),axis=0)*100
diff_acc_atmosphere = best_acc_atmosphere - mean_acc_atmosphere
```

# Changes in predictability depending on WR

```python
names_wr = ['WR1: AR','WR2: PT', 'WR3: GH', 'WR4: CUSH', 'WR5: Clima']
```

```python
for wr_temp in range(len(names_wr)):
    indices_seasons = {'Winter':[12,1,2],
                       'Spring':[3,4,5],
                       'Summer':[6,7,8],
                       'Fall':[9,10,11]}
    
    acc_persistence_seasons = {}
    for season in indices_seasons.keys():
        acc_persistence_temp = []
    
        for wk in range(1,9):
            df_temp = df_shifts[[f'week{0}',f'week{wk}']].dropna()
            df_temp = df_temp[df_temp[f'week{wk}']==wr_temp]
            df_temp = df_temp[(df_temp.index.month==indices_seasons[season][0])|\
                                        (df_temp.index.month==indices_seasons[season][1])|\
                                        (df_temp.index.month==indices_seasons[season][2])]
            acc_persistence_temp.append(accuracy_score(df_temp[f'week{0}'],df_temp[f'week{wk}']))
        acc_persistence_seasons[season] = acc_persistence_temp
    
    indices_seasons = {'Winter':[12,1,2],
                       'Spring':[3,4,5],
                       'Summer':[6,7,8],
                       'Fall':[9,10,11]}
    
    weeks = np.arange(1,9)
    for season,letter in zip(indices_seasons.keys(),['a) ','b) ','c) ','d) ']):
        for indices_component, component, label, color in zip([indices_atmosphere, indices_ocean, indices_land],
                                    ['atm','ocn','lnd'],
                                    ['atmosphere','ocean','land'],
                                    ['lightblue', 'darkblue', 'orange']):
            
            all_accs = []
            for ivar in range(len(indices_component)):
                name_temp = list_vars[indices_component[ivar]]
                acc_temp = []
                for week in weeks:
                    results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                              index_col=0,parse_dates=True)
                    results_temp = results_temp[(results_temp.index.month==indices_seasons[season][0])|\
                                                (results_temp.index.month==indices_seasons[season][1])|\
                                                (results_temp.index.month==indices_seasons[season][2])]

                    results_temp = results_temp[results_temp['y_true']==wr_temp]
                    acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
                all_accs.append(acc_temp)
            
            
            acc_temp = []
            for week in weeks:
                results_temp = pd.read_csv(f'{path_results}results_{component}_week{week}.csv',
                                                  index_col=0,parse_dates=True)
                results_temp = results_temp[(results_temp.index.month==indices_seasons[season][0])|\
                            (results_temp.index.month==indices_seasons[season][1])|\
                            (results_temp.index.month==indices_seasons[season][2])]
                
                results_temp = results_temp[results_temp['y_true']==wr_temp]
                acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
            # plt.plot(weeks,np.array(acc_temp)*100,label=f'All {label} variables',color=color,ls='--',alpha=0.4)
            all_accs.append(acc_temp)
            plt.plot(weeks,np.max(np.array(all_accs),axis=0)*100,label = f'Best {label} model',color=color)
            
            best_acc = np.max(np.array(all_accs),axis=0)*100
            mean_acc = np.mean(np.array(all_accs),axis=0)*100
            diffs = best_acc - mean_acc
            max_diff = 8
            min_diff = 0
            diffs = (diffs - min_diff)/(max_diff - min_diff)
            
            ind_max_acc = np.argmax(np.array(all_accs),axis=0)
            best_model = []
            for ind in ind_max_acc:
                if ind==len(all_accs)-1:
                    best_model.append('ALL')
                else:
                    best_model.append(' '.join(list_vars[indices_component[ind]].split('_')[:-1]))
            for i,j,d,diff in zip(weeks,np.max(np.array(all_accs),axis=0)*100,best_model,diffs):
                plt.scatter(i,j,color=color, s=diff*100)
                if component == 'lnd':
                    plt.text(i-0.32,j-5,s=d,color=color)
                else:
                    plt.text(i-0.28,j+4,s=d,color=color)
        
        plt.scatter(9,0,color='k',s=1*100,label='8% difference between best model and the mean\n accuracy of the models within each component')
        plt.scatter(9,0,color='k',s=0.125*100,label='1% difference between best model and the mean\n accuracy of the models within each component')
    
        plt.plot(weeks, np.array(acc_persistence_seasons[season][:])*100,
                 color='red',label='Persistence',ls='dashdot',lw=2)
        plt.xlim(0.6,8.4)
        plt.ylim(10,80)
        plt.grid(alpha=0.3)
        plt.ylabel('Accuracy [%]')
        plt.xlabel('Lead time [weeks]')
        
        plt.title(f'Accuracy from each component - {season} - {names_wr[wr_temp]}',
                  loc='left')
        # plt.legend(bbox_to_anchor=(1, 0.25))
        # plt.legend(fontsize=9)
    
        plt.savefig(f'figures_by_weather_regime/wr{wr_temp}_acc_3comps_{season}.png',bbox_inches='tight')
        # plt.show()
        plt.close('all')
```

```python
col_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_oni = pd.read_csv('https://psl.noaa.gov/data/correlation/oni.data', sep='\s+', skiprows=1, names=col_names) 
df_oni = df_oni.apply(pd.to_numeric, errors='coerce')
df_oni = df_oni.iloc[:-9]
# df_oni=df_oni.dropna(inplace=True)
# df_oni=df_oni.replace({-99.90 : np.NaN}, inplace=True)
stacked_series = df_oni.stack()
df_oni_stacked = stacked_series.to_frame()
df_oni_stacked.reset_index(inplace=True)
df_oni_stacked.columns = ['Year', 'Month', 'ONI']
df_oni_stacked['Date'] = pd.to_datetime(df_oni_stacked['Year'] + '-' + df_oni_stacked['Month'])
df_oni_stacked['Date'] = df_oni_stacked['Date'].dt.strftime('%Y-%m')
df_oni_stacked.set_index('Date', inplace=True)
df_oni_stacked.drop(['Year', 'Month'], axis=1, inplace=True)
df_oni_stacked.index = pd.to_datetime(df_oni_stacked.index)

serie_oni = []
for i in range(len(df_shifts.index)):
    date_temp = df_shifts.index[i]
    serie_oni.append(df_oni_stacked.loc[f'{date_temp.year}-{str(date_temp.month).zfill(2)}']\
            .values[0][0])
    
df_shifts['ONI'] = serie_oni
df_oni = df_shifts[['ONI']]
```

```python
names_states = ['Niño','Neutral','Niña']

acc_persistence_enso = {}
for state in names_states:
    acc_persistence_temp = []

    for wk in range(1,9):
        df_temp = df_shifts[[f'week{0}',f'week{wk}']].dropna()
        
        df_temp = pd.concat([df_temp,df_oni],axis=1).dropna()
        if state == 'Niño':
            df_temp = df_temp[df_temp['ONI']>=0.5]
        elif state == 'Neutral':
            df_temp = df_temp[(df_temp['ONI']>-0.5)&(df_temp['ONI']<0.5)]
        elif state == 'Niña':
            df_temp = df_temp[df_temp['ONI']<=-0.5]
        else:
            break
            
        acc_persistence_temp.append(accuracy_score(df_temp[f'week{0}'],df_temp[f'week{wk}']))
        
    acc_persistence_enso[state] = acc_persistence_temp
```

```python
weeks = np.arange(1,9)
for state in names_states:
    for indices_component, component, label, color in zip([indices_atmosphere, indices_ocean, indices_land],
                                ['atm','ocn','lnd'],
                                ['atmosphere','ocean','land'],
                                ['lightblue', 'darkblue', 'orange']):
        
        all_accs = []
        for ivar in range(len(indices_component)):
            name_temp = list_vars[indices_component[ivar]]
            acc_temp = []
            for week in weeks:
                results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                          index_col=0,parse_dates=True)
                
                results_temp = pd.concat([results_temp,df_oni],axis=1).dropna()
                if state == 'Niño':
                    results_temp = results_temp[results_temp['ONI']>=0.5]
                elif state == 'Neutral':
                    results_temp = results_temp[(results_temp['ONI']>-0.5)&(results_temp['ONI']<0.5)]
                elif state == 'Niña':
                    results_temp = results_temp[results_temp['ONI']<=-0.5]
                else:
                    break
                acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
            all_accs.append(acc_temp)
        
        
        acc_temp = []
        for week in weeks:
            results_temp = pd.read_csv(f'{path_results}results_{component}_week{week}.csv',
                                              index_col=0,parse_dates=True)
            results_temp = pd.concat([results_temp,df_oni],axis=1).dropna()
            if state == 'Niño':
                results_temp = results_temp[results_temp['ONI']>=0.5]
            elif state == 'Neutral':
                results_temp = results_temp[(results_temp['ONI']>-0.5)&(results_temp['ONI']<0.5)]
            elif state == 'Niña':
                results_temp = results_temp[results_temp['ONI']<=-0.5]
            else:
                break
            acc_temp.append(accuracy_score(results_temp['y_true'],results_temp['y_predicted']))
        # plt.plot(weeks,np.array(acc_temp)*100,label=f'All {label} variables',color=color,ls='--',alpha=0.4)
        all_accs.append(acc_temp)
        plt.plot(weeks,np.max(np.array(all_accs),axis=0)*100,label = f'Best {label} model',color=color)
        
        ind_max_acc = np.argmax(np.array(all_accs),axis=0)
        best_model = []
        for ind in ind_max_acc:
            if ind==len(all_accs)-1:
                best_model.append('ALL')
            else:
                best_model.append(' '.join(list_vars[indices_component[ind]].split('_')[:-1]))
        for i,j,d in zip(weeks,np.max(np.array(all_accs),axis=0)*100,best_model):
            plt.scatter(i,j,color=color)
            plt.text(i-0.25,j+0.27,s=d,color=color)
        
        plt.plot(weeks, np.array(acc_persistence_enso[state][:])*100,
             color='red',label='Persistence',ls='dashdot',lw=2)
    # plt.xlim(1,8)
    plt.ylim(19,40)
    plt.grid(alpha=0.3)
    plt.ylabel('Accuracy [%]')
    # plt.xlabel('Duration (weeks)')
    plt.title(f'Accuracy from each component - {state}: ',loc='left')
    plt.savefig(f'figures/acc_3comps_{state}.png'.replace('ñ','n'),bbox_inches='tight')
    # plt.legend(bbox_to_anchor=(1, 0.25))
    plt.show()
    plt.close('all')
```
