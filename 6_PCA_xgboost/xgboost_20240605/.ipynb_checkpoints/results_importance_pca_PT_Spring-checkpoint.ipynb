# Compute change in likelihood of WR=1 in spring

```python
df_wr_temp = df_shifts['week0'].dropna()
```

```python
index_season = (df_wr_temp.index.month==3)|\
            (df_wr_temp.index.month==4)|\
            (df_wr_temp.index.month==5)
```

```python
mean_freq_wr = len(df_wr_temp[(df_wr_temp==1)&index_season])/len(df_wr_temp[index_season])
```

```python
change_pc_positive = {}
change_pc_negative = {}

for week in weeks:
    # print(week)
    change_positive_week = []
    change_negative_week = []
    for ivar in indices_ordered:
        name_temp = list_vars[ivar]
        # print(name_temp)
        results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                  index_col=0,parse_dates=True)
        index_season = (results_temp.index.month==3)|\
            (results_temp.index.month==4)|\
            (results_temp.index.month==5)
        results_temp = results_temp[index_season]
        
        pcas = pd.read_csv(f'../figures_detrended_20240317/{name_temp}/PC_{name_temp}.csv',
                  index_col=0, parse_dates=True)
        index_season = (pcas.index.month==3)|\
            (pcas.index.month==4)|\
            (pcas.index.month==5)
        pcas = pcas[index_season]
        
        change_positive_week_var = []
        change_negative_week_var = []
        
        for ipc in range(len(pcas.keys())):
            # print(ipc)
            df_wr_temp = df_shifts[f'week{week}']
            pca_temp = pcas[f'{ipc}']
            perc_20 = np.percentile(pca_temp,20)
            perc_80 = np.percentile(pca_temp,80)
            df_pc_wr = pd.concat([pca_temp,df_shifts[f'week{week}']],axis=1).dropna()
            
            ## High values
            df_temp = df_pc_wr[(df_pc_wr[str(ipc)]>=perc_80)][f'week{week}']
            freq_temp = len(df_temp[df_temp==1])/len(df_temp)
            diff_freq_80 = freq_temp - mean_freq_wr

            ## Low values
            df_temp = df_pc_wr[(df_pc_wr[str(ipc)]<=perc_20)][f'week{week}']
            freq_temp = len(df_temp[df_temp==1])/len(df_temp)
            diff_freq_20 = freq_temp - mean_freq_wr
            
            change_positive_week_var.append(diff_freq_80) 
            change_negative_week_var.append(diff_freq_20)
        change_positive_week.append(change_positive_week_var)
        change_negative_week.append(change_negative_week_var)
    change_pc_positive[week] = np.array(change_positive_week)
    change_pc_negative[week] = np.array(change_negative_week)
    # aa
```

```python
threshold = 0.05
positive_freq_filtered = {}
for week in weeks:
    positive_freq_filtered[week] = copy.deepcopy(change_pc_positive[week])
    positive_freq_filtered[week][abs(change_pc_positive[week]<threshold)] = 0
    positive_freq_filtered[week][abs(change_pc_positive[week]>threshold)] = 1

negative_freq_filtered = {}
for week in weeks:
    negative_freq_filtered[week] = copy.deepcopy(change_pc_negative[week])
    negative_freq_filtered[week][abs(change_pc_negative[week]<=threshold)] = 0
    negative_freq_filtered[week][abs(change_pc_negative[week]>threshold)] = 1
```

```python
final_hatch = {}
for week in weeks:
    final_hatch[week] = positive_freq_filtered[week]+negative_freq_filtered[week]
    final_hatch[week][pc_mean_when_accurate[week]<10] = 0
```

# For each season

```python
indices_ordered = [23,8,21,22, 0, 1, 2, 3, 4, 5, 6, 7, 10, 12, 9, 13, 14, 15, 16, 17, 18, 19, 20] # atmosphere ocean land
weeks = np.arange(1,9)

indices_seasons = {'Winter':[12,1,2],
                   'Spring':[3,4,5],
                   'Summer':[6,7,8],
                   'Fall':[9,10,11]}

for season in indices_seasons.keys():
    
    pc_mean_when_accurate = {}
    for week in weeks:
    # for week in [1]:
        pc_when_accurate_week = []
        for ivar in indices_ordered:
            name_temp = list_vars[ivar]
            results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                      index_col=0,parse_dates=True)
            
            results_temp = results_temp[(results_temp.index.month==indices_seasons[season][0])|\
                (results_temp.index.month==indices_seasons[season][1])|\
                (results_temp.index.month==indices_seasons[season][2])]
            
            pcas = pd.read_csv(f'../figures_detrended_20240317/{name_temp}/PC_{name_temp}.csv',
                      index_col=0, parse_dates=True)
            pcas = pcas[(pcas.index.month==indices_seasons[season][0])|\
                (pcas.index.month==indices_seasons[season][1])|\
                (pcas.index.month==indices_seasons[season][2])]
            
            correct_indices = np.where(results_temp['y_true']==results_temp['y_predicted'])[0]
            df_correct = results_temp.iloc[correct_indices,0]
            acc_temp = len(correct_indices) / len(results_temp)
            
            pc_mag_all = pcas.abs().mean(axis=0).values
            pc_mag_correct = pcas.loc[df_correct.index].abs().mean(axis=0).values
            mag_change = pc_mag_correct - pc_mag_all
            mag_change_percentage = (100 * mag_change / pc_mag_all)# * acc_temp
            # norm_perc_change = mag_change_percentage * acc_temp
            pc_when_accurate_week.append(mag_change_percentage)
            # aaaa
        pc_mean_when_accurate[week] = np.array(pc_when_accurate_week)
        # pc_mean_when_accurate[week][pc_mean_when_accurate[week]<0] = np.nan
        # aaa
    
    df_wr_temp = df_shifts['week0'].dropna()
    mean_hist_wrs = np.histogram(df_wr_temp,bins=np.arange(-0.5,5,1))[0]
    mean_hist_wrs = mean_hist_wrs/np.sum(mean_hist_wrs)
    
    significance_pcs = {}
    for week in weeks:
        # print(week)
        significance_pcs_week = []
        for ivar in indices_ordered:
            name_temp = list_vars[ivar]
            # print(name_temp)
            results_temp = pd.read_csv(f'{path_results}results_{name_temp}_week{week}.csv',
                                      index_col=0,parse_dates=True)
            
            results_temp = results_temp[(results_temp.index.month==indices_seasons[season][0])|\
                (results_temp.index.month==indices_seasons[season][1])|\
                (results_temp.index.month==indices_seasons[season][2])]
            
            pcas = pd.read_csv(f'../figures_detrended_20240317/{name_temp}/PC_{name_temp}.csv',
                      index_col=0, parse_dates=True)
            
            pcas = pcas[(pcas.index.month==indices_seasons[season][0])|\
                (pcas.index.month==indices_seasons[season][1])|\
                (pcas.index.month==indices_seasons[season][2])]
            
            significance_pcs_week_var = []
            for ipc in range(len(pcas.keys())):
                # print(ipc)
                df_wr_temp = df_shifts[f'week{week}']
                pca_temp = pcas[f'{ipc}']
                perc_20 = np.percentile(pca_temp,30)
                perc_80 = np.percentile(pca_temp,70)
                df_pc_wr = pd.concat([pca_temp,df_shifts[f'week{week}']],axis=1).dropna()
    
                ## High values
                df_temp = df_pc_wr[(df_pc_wr[str(ipc)]>=perc_80)][f'week{week}']
                # hist_temp = np.histogram(df_temp,bins=np.arange(-0.5,5,1))[0]
                # hist_temp = hist_temp/np.sum(hist_temp)
                freq_table_A = df_wr_temp.value_counts().sort_index()##
                freq_table_B = df_temp.value_counts().sort_index()##
                observed = pd.concat([freq_table_A, freq_table_B], axis=1).fillna(0).values##
                _, p80, _, _ = chi2_contingency(observed)##
                
                ## Low values
                df_temp = df_pc_wr[(df_pc_wr[str(ipc)]<=perc_20)][f'week{week}']
                # hist_temp = np.histogram(df_temp,bins=np.arange(-0.5,5,1))[0]
                # hist_temp = hist_temp/np.sum(hist_temp)
                freq_table_A = df_wr_temp.value_counts().sort_index()##
                freq_table_B = df_temp.value_counts().sort_index()##
                observed = pd.concat([freq_table_A, freq_table_B], axis=1).fillna(0).values##
                _, p20, _, _ = chi2_contingency(observed)##
                # print (p20)
                if (p20<0.03)|(p80<0.03):
                    significance_pcs_week_var.append(1)
                else:
                    significance_pcs_week_var.append(0)
            significance_pcs_week.append(significance_pcs_week_var)
        significance_pcs[week] = np.array(significance_pcs_week)
        # aa
    significance_filtered = {}
    for week in weeks:
        significance_filtered[week] = copy.deepcopy(significance_pcs[week])
        significance_filtered[week][np.isfinite(pc_mean_when_accurate[week])==False] = 0
        significance_filtered[week][pc_mean_when_accurate[week]<=5] = 0
    for week in weeks:
        fig, ax1 = plt.subplots()
        # cm = ax1.pcolormesh(np.arange(0,20),np.arange(0,len(np.array(list_vars)[indices_ordered])),
        #                pc_mean_when_accurate[week],vmin = -0.1, vmax= 0.1,cmap='coolwarm')
        cm = ax1.pcolormesh(np.arange(0,20),np.array(list_vars)[indices_ordered],
                       pc_mean_when_accurate[week],vmin = 0, vmax= 12,cmap='YlOrRd')
        # Plot the meshgrid
        for i in range(len(np.array(list_vars)[indices_ordered])):
            for j in range(20):
                value = significance_pcs[week][i,j]
                if value == 1:
                    hatch_pattern = 'xx'
                    ax1.add_patch(plt.Rectangle((np.arange(0,20)[j]-0.5, np.arange(0,len(np.array(list_vars)[indices_ordered]))[i] - 0.5), 1, 1, 
                                                fill=False, hatch=hatch_pattern, color='gray',alpha=0.3))
                    
        # Plot the meshgrid
        for i in range(len(np.array(list_vars)[indices_ordered])):
            for j in range(20):
                value = significance_filtered[week][i,j]
                if value == 1:
                    hatch_pattern = 'xx'
                    ax1.add_patch(plt.Rectangle((np.arange(0,20)[j]-0.5, np.arange(0,len(np.array(list_vars)[indices_ordered]))[i] - 0.5), 1, 1, 
                                                fill=False, hatch=hatch_pattern, color='k',alpha=0.7))
                    
        plt.xticks(np.arange(0,20),fontsize=12)
        plt.yticks(fontsize=12)
    
        ax1.invert_yaxis()
        plt.title(f'{season} - Week {week}',fontsize=14)
        colorbar = plt.colorbar(cm)
        colorbar.set_label('[%]',fontsize=12)
        plt.savefig(f'heatmaps_seasons/heatmap_{season}_week{week}.png',bbox_inches='tight')
        plt.close('all')
```
