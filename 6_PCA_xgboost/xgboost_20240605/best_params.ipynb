```python
import json
import numpy as np
import xarray as xr
import pandas as pd
import copy
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import glob
```

```python
list_files = np.sort(glob.glob('best_params/*'))
```

```python
def read_json(path):
    # Open the JSON file
    with open(path, 'r') as file:
        # Load the data from the file
        data = json.load(file)
    return data
```

```python
dic_full = {}
keys = ['class_weights', 'colsample_bylevel', 'colsample_bytree', 'gamma', 'learning_rate', 
            'max_depth', 'min_child_weight', 'reg_alpha', 'reg_lambda', 'subsample']

for key in keys:
    dic_full[key] = []

for i in range(len(list_files)):
    data_temp = read_json(list_files[i])
    for j in range(len(data_temp)):
        dic_temp = data_temp[j]
        for key in keys:
            dic_full[key].append(dic_temp[key])
```

```python
# class weights
np.histogram(np.round(dic_full[keys[0]]))
```

```python
# for continuous variables
for i in range(1,len(keys)):
    plt.hist(dic_full[keys[i]],bins = 5)
    plt.title(keys[i])
    plt.show()
    plt.close()
```
