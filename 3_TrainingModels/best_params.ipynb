```python
import json
import numpy as np
import xarray as xr
import pandas as pd
import copy
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import glob
```

```python
list_files = np.sort(glob.glob('/glade/derecho/scratch/jhayron/Data4Predictability/ResultsXGBoost/ResultsHyperparamOptim_v20241226/*.json'))
```

```python
def read_json(path):
    # Open the JSON file
    with open(path, 'r') as file:
        # Load the data from the file
        data = json.load(file)
    return data
```

```python
dic_full = {}
keys = ['max_depth', 'min_child_weight', 'subsample', 'colsample_bytree', 'colsample_bylevel', 
            'log10_learning_rate', 'gamma', 'log10_reg_lambda', 'log10_reg_alpha', 'beta_class_weights']

for key in keys:
    dic_full[key] = []

for i in range(len(list_files)):
    data_temp = read_json(list_files[i])
    for key in keys:
        dic_full[key].append(data_temp[key])
```

```python
# for continuous variables
for i in range(len(keys)):
    plt.hist(dic_full[keys[i]],bins = 10)
    plt.title(keys[i])
    plt.show()
    plt.close()
```
