```python
for index in [17,18,19,20]:
# for index in [18,19,20]:
    ds = xr.open_dataset(list_datasets[index])
    ds = ds.compute()
    var_name = list(ds.data_vars.keys())[0]
    print(var_name)

    lat = ds.lat.values
    lon = ds.lon.values
    mask = np.zeros([len(lat),len(lon)])

    for ilat in range(len(lat)):
        # print(ilat)
        for ilon in range(len(lon)):
            serie_temp = ds.sel(lat=lat[ilat],lon=lon[ilon],method='nearest')[var_name]
            time_series = serie_temp.values
            window = 90
            forward_average = pd.Series(time_series).rolling(window=window, min_periods=1).mean().shift(-window + 1)
            backward_average = pd.Series(time_series).rolling(window,center=True,min_periods=1).mean().values
            diffs = abs(forward_average-backward_average)
            if (len(np.where((diffs/np.std(diffs))>=1)[0])/len(diffs))>0.3:
                mask[ilat,ilon] = 1
    path_masks = '/glade/work/jhayron/Data4Predictability/soil_masks/'
    path_save = f'{path_masks}{list_datasets[index].split('/')[-1][:-3]}.npy'
    np.save(path_save,mask)
```

```python
# Calculate absolute differences
differences = np.abs(np.diff(time_series))
# Define a threshold for identifying jumps
# threshold =  0.005 # Insert your desired threshold value here
diff = abs(np.diff(time_series))
threshold = np.percentile(diff,99.95)
# Find positions of steps above the threshold
step_positions = np.where(differences > threshold)[0] + 1  # Adding 1 to align with original indices

# Create a zero-valued offset array
offset_array = np.zeros_like(time_series)

# Insert differences of data points into offset array
offset_array[step_positions] = np.diff(time_series)[step_positions - 1]

# Form cumulative sum and subtract from original data
smoothed_data = time_series - np.cumsum(offset_array)

# 'smoothed_data' now contains the time series with steps removed
```

```python
file_temp = files_weekly_anoms[ifile]
dataset = xr.open_dataset(file_temp)
var_name = list(dataset.data_vars.keys())[0]
print(file_temp,var_name)
dataset = dataset.load()

name_var = file_temp.split('/')[-1][:-3]
### Find and drop useless (always zero, pixels)
flattened_da = dataset.stack(flat=('lat','lon')).transpose('time','flat')[var_name]
# nan_mask = flattened_da.isnull().all(dim='time')
nan_mask = flattened_da.isnull().any(dim='time')
where_nonan = np.where(nan_mask==False)[0]
flattened_nonan = flattened_da[:,where_nonan]
flattened_nonan = flattened_nonan.fillna(0)
# aaaa
# create pca object
pca_obj = PCA(20, whiten=True)##
pca_obj = pca_obj.fit(flattened_nonan)

# transform era5 data with pca
transformed_da = pca_obj.transform(flattened_nonan)
```

```python
plt.plot(flattened_nonan.time,transformed_da[:,3])
for iline in range(-5,0):
    plt.axvline(flattened_nonan.time[np.argsort(abs(np.diff(flattened_nonan[:,0])))[iline]].values,
               color='k')
plt.ylim(-1,1)
```

```python
plt.plot(flattened_nonan.time,transformed_da[:,6])
for iline in range(-3,0):
    plt.axvline(flattened_nonan.time[np.argsort(abs(np.diff(flattened_nonan[:,0])))[iline]].values,
               color='k')
plt.ylim(-1,1)
```
